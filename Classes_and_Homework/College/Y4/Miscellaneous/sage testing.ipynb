```python
u, v = var('u, v')
f_x(u, v) = u
f_y(u, v) = v
f_z(u, v) = u^2 + v^2
parametric_plot3d([f_x, f_y, f_z], (u, -2, 2), (v, -2, 2))
```

```python
u, v = var('u,v')
fx = (3+sin(v)+cos(u))*cos(2*v)
fy = (3+sin(v)+cos(u))*sin(2*v)
fz = sin(u)+2*cos(v)
parametric_plot3d([fx, fy, fz], (u, 0, 2*pi), (v, 0, 2*pi),
  frame=False, color="red")
```
